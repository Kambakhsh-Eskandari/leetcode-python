```python
class Node:
    def __init__(self, data = None, next=None):
        self.data = data
        self.next = next

class Linkedlist:
    def __init__(self):
        self.head = None
    
    def insert_at_head(self, data):
        node = Node(data, self.head)
        self.head = node
        
    def insert_at_end(self, data):
        if self.head is None:
            self.head = Node(data, None)
            return
        
        itr = self.head
        
        while itr.next:
            itr = itr.next
        itr.next = Node(data, None)
    
    def insert_list(self, List):
        self.head = None
        for val in List:
            self.insert_at_end(val)
            
    def get_length(self):
        itr = self.head
        count = 0
        
        while itr:
            count += 1
            itr = itr.next
        return count
    
    def remove_element(self, data):
        if self.head is None:
            print('list is empty')
            return
        itr = self.head
        if itr.next == data:
            itr.next = data.next
            return
        print('not found')
        return 
    
    def remove_at(self, index):
        if index > self.get_length():
            print('index exceeds the length')
            return
        
        if index == 0:
            self.head = self.head.next
            return
        
        count = 0
        itr = self.head
        while itr:
            if count == index - 1:
                itr.next = itr.next.next
                break
            
            itr = itr.next
            count += 1
        
    def insert_at(self, data, index):
        node = Node(data)
        if index == 0:
            self.insert_at_head(data)
            return
        
        if index > self.get_length():
            print('index exceeds the length')
            return
        
        itr = self.head
        count = 0
        while itr:
            if count == index - 1:
                node.next = itr.next
                itr.next = node
                break
            count += 1
            itr = itr.next


        
    
    def print(self):
        if self.head == None:
            print('list is empty')
            return
        
        itr = self.head
        lst = ''
        
        while itr:
            lst += str(itr.data) + "-->"
            itr = itr.next
        print(lst)
```
